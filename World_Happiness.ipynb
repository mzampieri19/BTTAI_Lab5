```python
# Find all nulls in the data frame
print(df.isnull().sum())
```

```python
# Replace all nulls with the mean of that feature
fill_in = ['Log GDP per capita', 
           'Social support', 
           'Healthy life expectancy at birth',
           'Freedom to make life choices',
           'Generosity',
           'Perceptions of corruption',
           'Positive affect',
           'Negative affect',
           'Confidence in national government',
           'Democratic Quality',
           'Delivery Quality',
           'GINI index (World Bank estimate)',
           'GINI index (World Bank estimate), average 2000-15',
           'gini of household income reported in Gallup, by wp5-year']

for i in fill_in: 
    df[i].fillna(df[i].mean(), inplace = True)

df.head(10)
```

```python
# Keep only the most recent entry for each country 

df = df.sort_values(by=['country', 'year'], ascending=[True, False])
df = df.drop_duplicates(subset=['country'], keep='first')
df.head(10)
```

```python
# Function to remove outliers from a collumn based on the IQR method 
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Using the function on the columns
columns = df.select_dtypes(include=['float64', 'int64']).columns

for i in columns: 
    remove_outliers(df, i)

df.head(10)
```

```python
# Remove year and country from the features 
df = df.drop(columns = ['country', 'year'])
```

```python
# Split data into X and y sets
y = df['Life Ladder']
X = df.drop(columns = 'Life Ladder', axis = 1)

# Create train and test split with 30% split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
# Begin training the model
model = LinearRegression()
model.fit(X_train, y_train)
prediction = model.predict(X_test)

# Get the weight and intercept of for the label 
print('Weight: ', model.coef_[0], ' intercept: ', model.intercept_)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, prediction)
r2 = r2_score(y_test, prediction)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

This result means that on average a prediction will be incorrect by .17 units, this is a good esitmation since the data for life ladder is between 0 and 1. A R2 value of .92 also signifies that the model is fairly accuarate. 

```python
# Set up the param grid
model = LinearRegression()

param_grid = {
    'fit_intercept': [True, False],
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score (neg_mean_squared_error): ", grid_search.best_score_)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred)
test_R2 = r2_score(y_test, y_pred)
print("Test MSE: ", test_mse)
print("Test R2: ", test_R2)
```

The best hyper parameter seems to be fit-intercept set to True, the MSE and R2 value does not seem to chanhe at all. This tells me that the original model is just as good as the optimized one.
